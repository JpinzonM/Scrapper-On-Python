```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd 
import numpy as np
```

```python
url="https://tienda.mercadolibre.com.co/lenovo#sellerid=234126796&banner_name=ADV_LENOVO_Smarter_BB_1"
results=requests.get(url)
```

```python
results
```

```python
soup = BeautifulSoup(results.text, "html.parser")
print(soup.prettify())
```

```python
Product = []
Price= []
Shipping= []
Imagen= []
```

```python
mercadolibre_div = soup.find_all('li', class_='results-item highlighted article grid item-info-height-169')
```

```python
print(mercadolibre_div)
```

```python
i =0 
for container in mercadolibre_div:
    name = container.h2.find('span', class_='main-title').text
    cost = container.find('div', class_='item__price').find('span', class_='price__fraction').text
    img = container.find('div', class_="images-viewer").find('img', attrs={"width":"284"})
    print("%i " %i, img)
    Product.append(name)
    Price.append(cost)
    i+=1
```

```python
i =0 
for container in mercadolibre_div:
    name = container.h2.find('span', class_='main-title').text
    cost = container.find('div', class_='item__price').find('span', class_='price__fraction').text
    img = container.find('div', class_="images-viewer").find('img', attrs={"width":"284"}).get("src") if container.find('div', class_="images-viewer").find('img', attrs={"width":"284"}).get("src") else container.find('div', class_="images-viewer").find('img', attrs={"width":"284"}).get("data-src")
    Product.append(name)
    Price.append(cost)
    Imagen.append(img)
```

```python
print(len(Product))
print(len(Price))
print(len(Imagen))
```

```python
df = pd.DataFrame({
    'name':Product,
    'cost': Price,
    'Imagen-Url':Imagen,
})
```
